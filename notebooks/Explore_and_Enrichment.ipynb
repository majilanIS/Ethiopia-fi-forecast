```python
#1. import necessary libraries and load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../data/raw/ethiopia_fi_unified_data.csv")
links = pd.read_csv("../data/raw/impact_sheet.csv")
codes = pd.read_csv("../data/raw/reference_codes.csv")

print("\n--- DATA SAMPLE ---")
print(data.head())

print("\n--- COLUMNS ---")
print(data.columns)

print("\n--- RECORD TYPES ---")
print(data["record_type"].value_counts())

print("\n--- PILLARS ---")
print(data["pillar"].value_counts(dropna=False))

print("\n--- SOURCE TYPES ---")
print(data["source_type"].value_counts())

print("\n--- CONFIDENCE ---")
print(data["confidence"].value_counts())

print("\n--- DATE RANGE ---")
print(data["observation_date"].min(), data["observation_date"].max())

print("\n--- INDICATORS ---")
print(data["indicator_code"].value_counts())

print("\n--- EVENTS ---")
print(data[data["record_type"]=="event"][["indicator","observation_date","category"]])

print("\n--- IMPACT LINKS ---")
print(links.head())

sns.set(style="whitegrid")
```

```python
#Pillar Coverage
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='record_type', palette='pastel', order=data['record_type'].value_counts().index)
plt.title("Records by Type")
plt.ylabel("Count")
plt.xlabel("Record Type")
plt.show()
```

```python
# Explore the Data

# Record type
print("\n--- Records by Type ---")
print(data['record_type'].value_counts())

# Pillar
print("\n--- Records by Pillar (NaN included) ---")
print(data['pillar'].value_counts(dropna=False))

# Source type
print("\n--- Records by Source Type ---")
print(data['source_type'].value_counts())

# Confidence
print("\n--- Records by Confidence ---")
print(data['confidence'].value_counts())
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))

sns.countplot(ax=axes[0,0], data=data, x='record_type', palette='pastel')
axes[0,0].set_title("Records by Type")

sns.countplot(ax=axes[0,1], data=data, x='pillar', palette='muted')
axes[0,1].set_title("Records by Pillar (NaN shown)")

sns.countplot(ax=axes[1,0], data=data, x='source_type', palette='coolwarm')
axes[1,0].set_title("Records by Source Type")

sns.countplot(ax=axes[1,1], data=data, x='confidence', palette='Set2')
axes[1,1].set_title("Records by Confidence")

plt.tight_layout()
plt.show()
```

```python
#List All Unique Indicators and Coverage
print("\n--- Unique Indicators ---")
print(data['indicator_code'].nunique())
print(data['indicator_code'].unique())

# Count coverage per indicator
indicator_counts = data['indicator_code'].value_counts()
print("\n--- Indicator Coverage ---")
print(indicator_counts)

# Optional: visualize top 10
plt.figure(figsize=(8,5))
sns.barplot(x=indicator_counts.head(10).values, y=indicator_counts.head(10).index, palette='viridis')
plt.title("Top 10 Indicators by Record Count")
plt.xlabel("Count")
plt.ylabel("Indicator Code")
plt.show()
```

```python
#Understand Which Events Are Cataloged and Their Dates
events = data[data['record_type']=='event'].sort_values('observation_date')
print("\n--- Events Sample ---")
print(events[['indicator','observation_date','category']].head(10))

# Number of events by category
print("\n--- Events by Category ---")
print(events['category'].value_counts())

# Event timeline visualization
plt.figure(figsize=(10,5))
sns.scatterplot(data=events, x='observation_date', y='category', s=100, color='orange')
plt.title("Events Over Time by Category")
plt.xlabel("Date")
plt.ylabel("Event Category")
plt.xticks(rotation=45)
plt.show()
```

```python
#Temporal Range of Observations
observations = data[data['record_type']=='observation']
print("\n--- Observation Date Range ---")
print("Earliest:", observations['observation_date'].min())
print("Latest:", observations['observation_date'].max())

print(observations['observation_date'].head(20))
print(observations['observation_date'].dtype)

observations['observation_date'] = pd.to_datetime(
    observations['observation_date'].astype(str).str.strip(), 
    errors='coerce', 
    dayfirst=True  # in case dates are like 31/12/2024
)

# Check how many failed
invalid_dates = observations[observations['observation_date'].isna()]
print("Rows with invalid dates:", len(invalid_dates))

observations = observations.dropna(subset=['observation_date'])
print("Data type now:", observations['observation_date'].dtype)
observations['year'] = observations['observation_date'].dt.year

# Convert Excel numeric dates to datetime
import datetime

def excel_date_to_datetime(val):
    try:
        val = float(val)
        return datetime.datetime(1899, 12, 30) + datetime.timedelta(days=val)
    except:
        return pd.NaT
```

```python
# Visualization
plt.figure(figsize=(10,4))
sns.histplot(observations['year'], bins=observations['year'].nunique(), color='skyblue')
plt.title("Observation Records Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
